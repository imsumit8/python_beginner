```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('vgsales.csv')
```

```python
data.head()
```

```python
data['Name'].value_counts()
```

```python
ih=data[data['Name']=='Ice Hockey']
sns.lineplot(data=ih,x='Year',y='NA_Sales',color='r')
plt.title('Ice hockey sales trend')
plt.show()
```

```python
sns.lineplot(data=data[data['Name']=='Baseball'],x='Year',y='NA_Sales',color='g')
plt.title('Base ball sales trend')
plt.show()
```

```python
sns.lineplot(data=ih,x='Year',y='NA_Sales',color='r',label='Ice Hockey')
sns.lineplot(data=data[data['Name']=='Baseball'],x='Year',y='NA_Sales',label='Base ball')
plt.legend(loc='lower left',frameon=False)
plt.show()
```

```python
sns.lineplot(data=ih,x='Year',y='NA_Sales',color='r',label='Ice Hockey')
plt.show()
sns.lineplot(data=data[data['Name']=='Baseball'],x='Year',y='NA_Sales',label='Base ball')
plt.show()
```

```python
top_3_pub=data['Publisher'].value_counts().iloc[0:3].index
```

```python
top_3_data=data[data['Publisher'].isin(top_3_pub)]
```

```python
top_3_data.head()
```

```python
sns.countplot(data=top_3_data,x='Publisher',hue='Genre')

plt.legend(loc='upper right')
plt.show()
```

```python
sns.boxplot(top_3_data,x='Publisher',y='NA_Sales')
```

```python
sns.boxplot(top_3_data,x='Genre',y='Global_Sales')
plt.title('Global sales per genre')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.subplot(1,3,1)
plt.scatter(x='NA_Sales',y='JP_Sales',data=top_3_data)
plt.subplot(1,3,3)
plt.scatter(x='NA_Sales',y='EU_Sales',data=top_3_data)
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.subplot(2,3,1)
plt.scatter(x='NA_Sales',y='JP_Sales',data=top_3_data,color='r')
plt.title('NA VS JAPAN')

plt.subplot(2,3,3)
plt.scatter(x='EU_Sales',y='JP_Sales',data=top_3_data,color='g')
plt.title('EU VS JAPAN')

plt.subplot(2,3,4)
plt.scatter(x='EU_Sales',y='NA_Sales',data=top_3_data,color='g')
plt.title('EU VS NA')

plt.subplot(2,3,6)
plt.scatter(x='NA_Sales',y='Global_Sales',data=top_3_data,color='g')
plt.title('NA VS GLOBAL')

plt.subplot(1,3,2)
sns.boxplot(top_3_data,x='Publisher',y='Global_Sales')
plt.title('EU VS JAPAN')
plt.show()
```
