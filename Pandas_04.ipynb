```python
movies = pd.read_csv('movies.csv', index_col=0)
directors = pd.read_csv('directors.csv', index_col=0)
```

```python
data=pd.read_csv('Pfizer_1.csv')
```

```python
data.head()
```

```python
data_melt=pd.melt(data,id_vars=['Date','Drug_Name','Parameter'],value_name='Reading',var_name='Time')
data_melt
```

```python
data_tidy=pd.pivot(data_melt,index=['Date','Drug_Name','Time'],columns='Parameter',values='Reading').reset_index()
```

```python
data_tidy.head()
```

```python
data_tidy.isnull().sum()
```

```python
def temp_mean(x):
  x['Temp_avg']=x['Temperature'].mean().round()
  return x
```

```python
data_tidy=data_tidy.groupby('Drug_Name',group_keys=False).apply(temp_mean)
```

```python
data_tidy['Temperature'].fillna(data_tidy['Temp_avg'],inplace=True)
```
