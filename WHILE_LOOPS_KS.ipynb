```python
print('''

Question 1 -
Take an integer input from the user and print alternate numbers from from 1 to input.
Input will be a positive number > 1. INPUT - 7
OUTPUT - 1 3 5 7 INPUT - 4
OUTPUT - 1 3'''
)
```

```python
number=int(input())
i=1
while i<=number:
  print(i,end=" ")
  i+=2
```

```python
print('''Question 1.1 -
Take an input from the user and print alternate numbers from from 1 to input.
If number is > 1 -> Print as required.
Else print "Number is smaller than 1!" INPUT - 7
OUTPUT - 1 3 5 7 INPUT - -5
OUTPUT - Number is smaller than 1!
''')
```

```python
number=int(input())
i=1
if number>=1:
  while i<=number:

    print(i,end=" ")

    i+=2
else:
  print("Number is smaller than 1!")
```

```python
print('''
Question 2 -
Take two integer inputs from user for start and end integer values.
Print every third number starting at start till end. (BOTH START AND END TO BE INCLUDED.) Input :
2
30
2 5 8 11 14 17 20 23 26 29
''')
```

```python
start=int(input())
end=int(input())
while start<=end:
  print(start,end=" ")
  start+=3
```

```python
print('''Question 3 -
Take an input from user for start and end values.
Print all even numbers between them. (start and end included) INPUT -
3
7 OUTPUT - 4 6 '''
)
```

```python
start=int(input())
end=int(input())

if start%2==0:
  final_start=start
else:
  final_start=start+1

while final_start<=end:
  print(final_start,end=" ")
  final_start+=2
```

```python
print("""Password is 567 Write a loop in python which runs from 0 to 999 and prints the password after finding it.
Print - "Trying <number>" for every number
Print - "Password found - number" when password is found""")
```

```python
password=567
i=0
while i<=999:
  print("typing",i)
  if i==password:
    print("password found",i)

  i+=1
```

```python
print('''Take a number input from user.
Print its multiplication table till 10th multiple. INPUT - 7
OUTPUT - 7 14 21 28 35 42 49 56 63 70
''')
```
